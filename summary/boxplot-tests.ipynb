## Feature Extraction
### Countvectorizer & TF - IDF
We use two methods to do the feature extraction on our review text data. First one is countvectorizer which counts only the frequency of words in the text. In addition to count the frequency of a word, tfidfvectorizer also considers the number of all texts containing the word. It can reduce the impact of frequently appearing meaningless words and explore more meaningful features. By contrast, the more text items there are, the more significant Tfidf will be.

## Correlation test
### In TF-IDF step, from distribution plots, we can see 'bread' and 'soup' / 'bacon' and 'dessert' / 'tea' and 'dessert' seem to have a similar trend among different stars' reviews.  To test whether there exists correlation between those two food items, we use Spearman's method to do the correlation test.
#### H0: the value of the association measure = 0, which means the two samples are uncorrelated.
The table shows the TF - IDF score of the food items.

| |star1 |star2 |star3 |star4 |star5|
|-----|----|-----|------|-----|------|
|bread|0.02457| 0.03506| 0.03696| 0.0396| 0.03193|
|soup|0.01836| 0.02688| 0.02985| 0.03124| 0.024|

#### Spearman's rank correlation rho:  S = 4.4409e-15, p-value = 0.01667
alternative hypothesis: true rho is not equal to 0 /// sample estimates:  rho    1 


| |star1 |star2 |star3 |star4 |star5|
|-----|----|-----|------|-----|------|
|bacon|0.03468| 0.04667| 0.05561| 0.06337| 0.05079|
|dessert|0.01947| 0.03348| 0.0485| 0.05582| 0.04688|

#### Spearman's rank correlation rho:  S = 4.4409e-15, p-value = 0.01667
alternative hypothesis: true rho is not equal to 0 /// sample estimates:  rho   1 


| |star1 |star2 |star3 |star4 |star5|
|-----|----|-----|------|-----|------|
|tea|0.00814| 0.01107| 0.01167| 0.01423| 0.01293|
|dessert|0.01947| 0.03348| 0.0485| 0.05582| 0.04688|

#### Spearman's rank correlation rho:  S = 2, p-value = 0.08333
alternative hypothesis: true rho is not equal to 0 /// sample estimates:  rho   0.9


#### Summary of Correlation test: 
All of the 3 hypothses reject H0 at 90% significance level, so these three pairs of words are correlated with each other respectively. These food items do have a similar trend.
